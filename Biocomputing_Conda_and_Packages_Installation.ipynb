```python
%%bash
# Initialize conda in the current shell with the correct installation path
eval "$(/root/miniconda3/bin/conda shell.bash hook)"

# Activate the base environment
conda activate base

# Verify conda installation
conda --version
```

```python
%%bash
# Initialize conda in the current shell
eval "$(/root/miniconda3/bin/conda shell.bash hook)"

# Activate the base environment (optional, but good practice)
conda activate base

# Accept Conda Terms of Service for required channels
conda tos accept --override-channels --channel https://repo.anaconda.com/pkgs/main
conda tos accept --override-channels --channel https://repo.anaconda.com/pkgs/r

# Create a conda environment named 'funtools' with Python 3.10
conda create -n funtools python=3.10 -y
```

```python
%%bash
# Initialize conda in the current shell
eval "$(/root/miniconda3/bin/conda shell.bash hook)"

#Activate the funtools environment
conda activate funtools
```
