```python
#Print the sequence length of the .gbk file. (Use the LOCUS tag in the first line)
#Print the source organism of the .gbk file. (Use the SOURCE tag in the first line)
import re

sequence_length = None
source_organism = None

with open('wildtype.gbk', 'r') as f:
    for line in f:
        if line.startswith('LOCUS'):
            match = re.search(r'\s+(\d+)\s+bp', line)
            if match:
                sequence_length = match.group(1)
        elif line.startswith('SOURCE'):
            source_organism = line.replace('SOURCE', '').strip()

        if sequence_length is not None and source_organism is not None:
            break # Stop reading once both are found

print(f"Sequence Length: {sequence_length} bp")
print(f"Source Organism: {source_organism}")
```
