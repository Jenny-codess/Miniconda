```python
%%bash

# 1. Download Miniconda installer
curl -LO "https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh"

# 2. Make the installer executable
chmod +x Miniconda3-latest-Linux-x86_64.sh

# 3. Run the installer in silent mode
# -b for batch mode (non-interactive)
# -p specifies the installation path
./Miniconda3-latest-Linux-x86_64.sh -b -p /root/miniconda3

# 4. Remove the installer script (optional cleanup)
rm Miniconda3-latest-Linux-x86_64.sh

# 5. Initialize conda for the current bash shell session
# This command modifies the current shell's PATH to include conda,
# allowing subsequent conda commands to be recognized.
# Note: For conda to be available in new shell sessions (e.g., other %%bash cells),
# you might need to re-run this eval command or configure persistent shell initialization.
eval "$(/root/miniconda3/bin/conda shell.bash hook)"

# 6. Activate the base environment
# (Often automatically active after initialization, but explicit activation is good practice)
conda activate base

# 7. Verify conda installation
echo "\nMiniconda installation verified:"
conda --version
```
