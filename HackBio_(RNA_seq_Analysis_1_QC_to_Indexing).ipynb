# Trimming

```python
%%bash
# This script is used to trim and echo files for our RNA-seq data for C. elegans

# Initialize conda for the current bash shell session
eval "$(/root/miniconda3/bin/conda shell.bash hook)"

# Activate the funtools environment
conda activate funtools

# create directories for output
mkdir -p trim

for filename in clean_files/*.fq; do
    base=$(basename "$filename" .fq)    #extract the basename of each files
    echo fastp -i "$filename" -o "trim/${base}_trim.fq" -h "trim/${base}_report.html"
done
```

Create Transcriptome Maps

```python
%%bash
# Initialize conda for the current bash shell session
eval "$(/root/miniconda3/bin/conda shell.bash hook)"

# Activate the funtools environment
conda activate funtools

# Navigate back to the main content directory to ensure correct path resolution
cd /content

# create directories for output
mkdir -p mapped_reads

#script for mapping reads
for infile in trim/*.fq ; do
	outfile=$(basename "$infile" .fq)    #extract the basename for each files in trim directory apart from '.fq'
  STAR --genomeDir reference_genome/genomeIndex --readFilesIn $infile --outFileNamePrefix mapped_reads/$outfile --outSAMtype BAM SortedByCoordinate --outSAMattributes All
done
```

# Run featureCounts to count transcript

##### -O: count reads overlapping multiple features
##### -t: feature type to count (gene)
##### -g: attribute to group features (ID)
##### -a: annotation file (GFF3)
##### -o: output file


```python
#run featureCounts to count transcripts
%%bash
# Initialize conda for the current bash shell session
eval "$(/root/miniconda3/bin/conda shell.bash hook)"

# Activate the funtools environment
conda activate funtools

featureCounts -O -t gene -g ID -a C_elegans.gff3 -o count.txt ../IGV/*.bam
```
